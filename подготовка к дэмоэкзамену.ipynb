```python
#@title func {display-mode:"form"}

def get_line_from_file(file):
  RecordID = None
  Survival = None
  In_hospital_death = None
  Albumin = []
  ALP = []
  ALT = []
  AST = []
  Bilirubin = []
  BUN = []
  Cholesterol = []
  Creatinine = []
  DiasABP = []
  FiO2 = []
  GCS = []
  Glucose = []
  HCO3 = []
  HCT = []
  HR = []
  K = []
  Lactate = []
  Mg = []
  MAP = []
  MechVent = []
  Na = []
  NIDiasABP = []
  NIMAP = []
  NISysABP = []
  PaCO2 = []
  PaO2 = []
  pH = []
  Platelets = []
  RespRate = []
  SaO2 = []
  SysABP = []
  Temp = []
  TropI = []
  TropT = []
  Urine = []
  WBC = []
  Weight = []

  with open(file, 'r') as f:
    lines = f.readlines()
    for line in lines:
      line_values = line.split(',')
      if line_values[1] == 'RecordID':
        RecordID = int(line_values[2])
      if line_values[1] == 'Albumin':
        Albumin.append(float(line_values[2]))
      if line_values[1] == 'ALP':
        ALP.append(float(line_values[2]))
      if line_values[1] == 'ALT':
        ALT.append(float(line_values[2]))
      if line_values[1] == 'AST':
        AST.append(float(line_values[2]))
      if line_values[1] == 'Bilirubin':
        Bilirubin.append(float(line_values[2]))
      if line_values[1] == 'BUN':
        BUN.append(float(line_values[2]))
      if line_values[1] == 'Cholesterol':
        Cholesterol.append(float(line_values[2]))
      if line_values[1] == 'Creatinine':
        Creatinine.append(float(line_values[2]))
      if line_values[1] == 'DiasABP':
        DiasABP.append(float(line_values[2]))
      if line_values[1] == 'FiO2':
        FiO2.append(float(line_values[2]))
      if line_values[1] == 'GCS':
        GCS.append(float(line_values[2]))
      if line_values[1] == 'Glucose':
        Glucose.append(float(line_values[2]))
      if line_values[1] == 'HCO3':
        HCO3.append(float(line_values[2]))
      if line_values[1] == 'HCT':
        HCT.append(float(line_values[2]))
      if line_values[1] == 'HR':
        HR.append(float(line_values[2]))
      if line_values[1] == 'K':
        K.append(float(line_values[2]))
      if line_values[1] == 'Lactate':
        Lactate.append(float(line_values[2]))
      if line_values[1] == 'Mg':
        Mg.append(float(line_values[2]))
      if line_values[1] == 'MAP':
        MAP.append(float(line_values[2]))
      if line_values[1] == 'MechVent':
        MechVent.append(float(line_values[2]))
      if line_values[1] == 'Na':
        Na.append(float(line_values[2]))
      if line_values[1] == 'NIDiasABP':
        NIDiasABP.append(float(line_values[2]))
      if line_values[1] == 'NIMAP':
        NIMAP.append(float(line_values[2]))
      if line_values[1] == 'NISysABP':
        NISysABP.append(float(line_values[2]))
      if line_values[1] == 'PaCO2':
        PaCO2.append(float(line_values[2]))
      if line_values[1] == 'PaO2':
        PaO2.append(float(line_values[2]))
      if line_values[1] == 'pH':
        pH.append(float(line_values[2]))
      if line_values[1] == 'Platelets':
        Platelets.append(float(line_values[2]))
      if line_values[1] == 'RespRate':
        RespRate.append(float(line_values[2]))
      if line_values[1] == 'SaO2':
        SaO2.append(float(line_values[2]))
      if line_values[1] == 'SysABP':
        SysABP.append(float(line_values[2]))
      if line_values[1] == 'Temp':
        Temp.append(float(line_values[2]))
      if line_values[1] == 'TropI':
        TropI.append(float(line_values[2]))
      if line_values[1] == 'TropT':
        TropT.append(float(line_values[2]))
      if line_values[1] == 'Urine':
        Urine.append(float(line_values[2]))
      if line_values[1] == 'WBC':
        WBC.append(float(line_values[2]))
      if line_values[1] == 'Weight':
        Weight.append(float(line_values[2]))

  result = [RecordID, Survival, In_hospital_death, Albumin, ALP, ALT, AST, Bilirubin, BUN, Cholesterol, Creatinine, DiasABP, FiO2, GCS, Glucose, HCO3, HCT, HR, K, Lactate, Mg, MAP, MechVent, Na, NIDiasABP, NIMAP, NISysABP, PaCO2, PaO2, pH, Platelets, RespRate, SaO2, SysABP, Temp, TropI, TropT, Urine, WBC, Weight]
  for i in range(len(result)):
    if type(result[i]) == list:
      result[i] = np.mean(result[i])


  return result
```
